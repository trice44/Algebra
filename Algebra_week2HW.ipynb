```python
#get data and read first lines to identify variables
import pandas as pd

Location = "datasets/algebradata.csv"

df = pd.read_csv(Location)
df.head()
```

```python
#get total count of alegebra students
total_students = df['Grade'].count()
print total_students
```

```python
#count the students with failing grades. 
df.loc[df['Grade'].str.contains('F')==True].count()
```

```python
#get a count of all participants with passing grade and use to create percentage
number_pass = df.loc[df['Grade'].str.contains('F')==False].count()
percent_pass = number_pass/total_students
print percent_pass




```
